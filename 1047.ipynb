Leia a hora inicial, minuto inicial, hora final e minuto final de um jogo. A seguir calcule a duração do jogo.

Obs: O jogo tem duração mínima de um (1) minuto e duração máxima de 24 horas.

Entrada
Quatro números inteiros representando a hora de início e fim do jogo.

Saída
Mostre a seguinte mensagem: “O JOGO DUROU XXX HORA(S) E YYY MINUTO(S)” .
24 10 24 9



```python
h_ini,m_ini,h_fin,m_fin = map(int,input().split(' '))
h_res = abs(h_ini-h_fin) if h_ini < h_fin else (24 - h_ini)+h_fin
m_res = abs(m_ini-m_fin) if m_ini < m_fin else (60 - m_ini)+m_fin
if m_ini == m_fin:
    m_res = 0
if h_ini == h_fin and m_ini > m_fin:
    h_res -= 1
elif h_ini == h_fin and m_ini < m_fin:
    h_res = 0
elif h_ini > h_fin and m_ini > m_fin:
    h_res -= 1
elif h_ini < h_fin and m_ini > m_fin:
    h_res -= 1
elif m_ini == 60:
    m_res = m_fin
elif h_res == 1 and m_ini > m_fin:
    h_res = 0

print(f"O JOGO DUROU {h_res} HORA(S) E {m_res} MINUTO(S)")

#if tempo_resto <= 24 and m_ini >= 1:
 #   print("O JOGO DUROU 24 HORA(S)" if a==b==0 else f"O JOGO DUROU {tempo_resto} HORA(S)")
#else:
 #   print(f"O JOGO DUROU {tempo_resto} HORA(S)")
```
