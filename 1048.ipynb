```python
valor = float(input())
reajuste = []
if valor <= 400:
    reajuste = [(valor / 100) * 15]
    reajuste += [15]
elif valor <= 800:
    reajuste = [(valor / 100) * 12]
    reajuste += [12]
elif valor <= 1200:
    reajuste = [(valor / 100) * 10]
    reajuste += [10]
elif valor <= 2000:
    reajuste = [(valor / 100) * 7]
    reajuste += [7]
else:
    reajuste = [(valor / 100) * 4]
    reajuste += [4]
print(f"""Novo salario: {valor+reajuste[0]:.2f}
Reajuste ganho: {reajuste[0]:.2f}
Em percentual: {reajuste[1]} %""")
```
